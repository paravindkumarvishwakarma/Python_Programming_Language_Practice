Que-91. Using arithmetic operators, ask the user for two integers and describe what +, -, *, /, //, %, ** would compute for them.

```python
#Ask for the user input
a = int(input("Enter first number = "))
b = int(input("Enter Second number = "))

#Performing operations
print("\nresults")
print(f"a + b (addition)              = {a + b}")
print(f"a - b (subtraction)           = {a - b}")
print(f"a * b (multiplication)        = {a * b}")

#Handling division by zero safely
if b != 0:
    print(f"a / b  (true division)    = {a / b}")   # Float Result
    print(f"a // b (Floor division)   = {a // b}")  # Integer quotient
    print(f"a % b  (modulus)          = {a % b}")   # Remainder
else:
    print(f"a / b  (true division)    = undefined (division by zero")
    print(f"a // b (Floor division)   = undefined (division by zero")
    print(f"a % b  (modulus)          = undefined (division by zero")
```

Que-106. Demonstrate short-circuit behavior: explain what happens when evaluating (False and expr) and (True or expr).

```python
#1) (False and expr)
def side_effect():
    print("expr evaluated")
    return True

result = (False and side_effect())
print(result)

#2) (True or expr)
def side_effect():
    print("expr evaluated")
    return False

result = (True or side_effect())
print(result)
```
