Que-31. Take a number as input and print the sum of its digits.

```python
print("Que-31. Take a number as input and print the sum of its digits.")
num_int = int(input("Enter a number = "))

sum_digits = 0
while num_int > 0:
    digit = num_int % 10
    sum_digits += digit
    num_int = num_int // 10

print("Sum of digits =", sum_digits)
```

Que-33. Take a number as input and check if it is a prime number.

```python
num_Prime = int(input("Enter a number = "))

if num_Prime <= 1:
    print("False")
elif num_Prime == 2:
    print("True")
elif num_Prime % 2 == 0:
    print("False")
else:
    is_prime = True
    for i in range(3, int(num_Prime**0.5) + 1, 2):
        if num_Prime % i == 0:
            is_prime = False
            break
    print("True" if is_prime else "False")

print("\n")
```
